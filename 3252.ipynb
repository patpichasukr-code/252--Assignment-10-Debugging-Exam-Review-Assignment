```python
def average_by_key(rows):
    total = 0
    count = 0
    for r in rows:
        if "score" in r:          # check key exists
            total += r["score"]   # fix syntax error
            count += 1
    if count == 0:
        return 0                 # handle empty or no-score input
    return total / count


print(average_by_key([{"score": 5}, {"score": 7}, {"x": 1}]))  # Expected Output: 6.0
print(average_by_key([]))   # Expected Output: 0
```

```python
def even_numbers(nums):
    evens =[]
    for n in nums:
        if n % 2 == 0:
            evens.append(n)
    return evens

print(even_numbers([1,2,3,4,5,6])) # Expected Output: [2, 4, 6]
```

```python
class Artist:
    def __init__(self, name, fee):
        self.name = name
        self.__fee = fee #private attribute, do not change this.

    def add_charges(self, additional_charges):
        self.fee = self.__fee + additional_charges

a1 = Artist("Taylor Swift", 50)
a2 = Artist("Lisa", 65)
a1.add_charges(20)
a2.add_charges(20)
print( a1.fee) # Expected Output: 70
print( a2.fee) # Expected Output: 85
```
